### ✅ KPI란? (Key Performance Indicator)

KPI는 **핵심성과지표**로,  
조직이나 팀이 정한 **목표 달성 여부를 측정하기 위한 핵심 지표**입니다.

#### 📌 예시
- 온라인 쇼핑몰: `재구매율`, `장바구니 이탈률`
- 헬스장: `월간 신규 가입자 수`
- 카드사: `총 이용금액 증가율`, `혜택 수혜율`

---

### ✅ 좋은 KPI의 조건
| 조건 | 설명 |
|------|------|
| 🎯 측정 가능성 | 수치로 나타낼 수 있어야 함 |
| 🎯 명확성 | 지표가 무엇을 의미하는지 분명해야 함 |
| 🎯 목표 연관성 | 비즈니스 목표와 직접 연결되어야 함 |
| 🎯 핵심성 | 많지 않게, 정말 중요한 지표만 선택 |

---

### ✅ 데이터 기반 KPI 후보 (1차 선정)

| KPI 후보 | 관련 컬럼명 | 의미 요약 |
|----------|-------------|------------|
| **이용건수/금액 증감률 (활동성)** | `증감율_이용금액_신용_전월`, `증감율_이용금액_일시불_전월`, `증감율_이용건수_카드론_분기` 등 | 최근 이용량의 증가/감소 흐름을 통해 고객의 활동성 추정 |
| **카드 유형별 사용 증감률** | `증감율_이용건수_체크_전월`, `증감율_이용금액_카드론_분기` 등 | 고객의 결제 방식 선호도 파악 (신용 vs 체크 vs 할부 등) |
| **한도 소진율** | `잔액_신판평균한도소진율_r6m`, `잔액_신판ca최대한도소진율_r3m` 등 | 고객의 카드 사용 여력 또는 한도 소비 성향 분석 |
| **변동률 기반 소비 안정성** | `변동률_할부평잔`, `변동률_카드론평잔` 등 | 소비 패턴의 일관성 또는 불안정성 탐지 |
| **잔액 변동성** | `변동률_잔액_B1M`, `변동률_잔액_CA_B1M` 등 | 남은 잔액이 급격히 변하는 경우, 소비 습관 변화의 지표 |
| **혜택 수혜율** | `혜택수혜율_R3M`, `혜택수혜율_B0M` | 리워드/이벤트 혜택 활용도 → 혜택 중심 소비인지 파악 가능 |

---

### ✅ KPI는 세그먼트 분석에 어떻게 활용될 수 있을까?

KPI는 **각 세그먼트를 구별하는 기준점**이 될 수 있습니다.

| 예시 | 설명 |
|------|------|
| VIP 고객 | `혜택수혜율`, `이용금액 증감률`, `한도 소진율`이 높은 경향 |
| 이탈 고객 | `이용건수 증감률`이 크게 하락, `변동률`이 높음 |
| 신규 고객 | 특정 항목이 `0 또는 급증`하며, 사용 패턴이 아직 정착되지 않음 |

📊 이처럼 KPI를 기준으로 boxplot, 평균 비교, 상관분석을 통해 **세그먼트의 특성 구분**이 가능하며,  
간단한 분류모델로 **KPI가 예측력 있는지**도 확인해볼 수 있습니다.

# KPI 분석 관련 질의 응답 정리

---

## 1. 증감율 컬럼에 음수 값이 있어도 괜찮은가?

✅ **괜찮다.**

- 증감율은 기준 대비 변화량을 나타내는 지표이므로,
  - **음수**는 감소,
  - **양수**는 증가를 의미한다.
- 이탈 고객은 사용량이 줄어들면서 자연스럽게 음수가 될 수 있음.
- 반대로 VIP 고객은 사용량이 증가하면서 양수일 가능성이 큼.
- 따라서 **음수 값 그대로 분석에 사용 가능.**

---

## 2. 그래프는 유의미하지 않았지만, 표 상으로는 어떤가?

📌 표를 보면 **세그먼트 A/B/C와 D/E 간의 분리 가능성**이 일부 보임.

예시:
- `증감율_이용금액_신판_전월`: A~C가 높고, E는 낮음
- `변동률_CA평잔`, `변동률_카드론평잔`: A~C 높고, E 낮음

⚠️ 다만 전반적인 편차는 크지 않아 **단독 지표로는 부족**할 수 있음.

---

## 3. EDA 중 유의미했던 컬럼을 KPI로 사용해도 되나?

✅ **그렇다. 매우 자연스러운 방식.**

- KPI는 꼭 통계적으로 유의하지 않아도 됨.
- 다음 기준에 부합하면 충분히 사용 가능:
  - 비즈니스적으로 **성과를 나타냄**
  - **이탈 고객의 행동을 설명함**
  - **우수 고객의 특징을 반영함**

---

## 4. KPI 지표로 선택되는 컬럼 수에 제한이 있나?

📌 일반적인 실무 기준:
- **3~5개 이내**로 설정하는 것이 바람직.
- 이유:
  - KPI는 핵심 지표이므로 너무 많으면 초점이 흐려짐
  - 보고서나 대시보드에 명확히 표현하기 쉬움
  - 클러스터별 비교나 모니터링이 용이함

👉 단, 지금은 **분석 초기 단계**이므로
- 후보를 넉넉히 뽑고,
- 이후 Feature Importance 등을 통해 줄이는 방식 추천.

### 가중치를 다시 둘 수 있는 방법 재고
- 세그먼트 간 평균의 차이를 기준으로 KPI 컬럼별 가중치를 부여하는 방법

```python
import pandas as pd

# KPI 후보 컬럼들 (예시 리스트로 작성해둡니다)
kpi_cols = [
    '증감율_이용건수_할부_전월',
    '증감율_이용건수_CA_전월',
    '증감율_이용건수_체크_전월',
    '증감율_이용금액_신용_전월',
    '증감율_이용금액_신판_전월',
    '증감율_이용금액_일시불_전월',
    '증감율_이용금액_할부_전월',
    '증감율_이용건수_신판_분기',
    '증감율_이용건수_일시불_분기',
    '증감율_이용금액_신용_분기',
    '증감율_이용금액_체크_분기',
    '변동률_RV일시불평잔',
    '변동률_할부평잔',
    '변동률_RVCA평잔',
    '변동률_카드론평잔'
]

# 1. Segment 별 평균 구하기
segment_means = result_df.groupby("Segment")[kpi_cols].mean()

# 2. 각 컬럼별 (최대 - 최소) 범위 계산
mean_range = segment_means.max() - segment_means.min()

# 3. 전체 총합으로 나눠서 가중치 계산
weights = mean_range / mean_range.sum()

# 4. 각 row에 가중치를 곱해서 KPI Score 계산
result_df['KPI_Score'] = result_df[kpi_cols].mul(weights, axis=1).sum(axis=1)

# 결과 확인 (상위 몇 개만 출력)
print(result_df[['Segment', 'KPI_Score']].head())
```

### 각 세그먼트 별로 동일한 샘플 수를 추출

```python
import pandas as pd

# 세그먼트별 최소 샘플 수 확인
min_count = result_df['Segment'].value_counts().min()
print(f"최소 샘플 수: {min_count}")

# 각 세그먼트별로 동일하게 샘플링
balanced_df = (
    result_df.groupby('Segment')
    .apply(lambda x: x.sample(n=min_count, random_state=42))
    .reset_index(drop=True)
)
```

2. **세그먼트별 분포 특징**
   - **D, E** 세그먼트는 평균 KPI Score가 상대적으로 높고, **상위 이상치(Outlier)**도 자주 발생함.
   - 반면 **B**는 중앙값이 0보다 낮으며 전체적으로 성과 지표가 낮은 분포를 보임.

   - 활동성이 높았던 A, B 세그먼트가 KPI Score에서는 상위가 아닐 수 있다는 점은,
   - 단순 활동성보다 **변동성, 증가율, 리스크 요인** 등이 KPI에 더 큰 영향을 주었음을 시사함.

4. **시사점**
   - KPI는 단순 평균보다는 표준화와 샘플링을 고려해 분석할 필요가 있음.
   - 이 결과는 이후 세그먼트 리포트 또는 마케팅 전략 수립에 기초 지표로 사용될 수 있음.


```python
import scipy.stats as stats

# 세그먼트별 KPI Score 그룹 나누기
groups = [group["KPI_Score"].values for name, group in result_df.groupby("Segment")]

# 일원분산분석 (One-Way ANOVA)
f_stat, p_val = stats.f_oneway(*groups)

print(f"F-statistic: {f_stat:.4f}")
print(f"p-value: {p_val:.4f}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=result_df['KPI_Score'],
                          groups=result_df['Segment'],
                          alpha=0.05)

print(tukey)
```

- **A, B 세그먼트 → 리텐션 강화 필요**, KPI가 낮은 고객군.
- **C, D, E 세그먼트 → 우수 그룹**, 유지 혹은 프리미엄 전략 고려.
- 이 결과를 기반으로 KPI 기준 **세그먼트 재정의** 또는 **마케팅 타겟팅** 가능.



## 📊 KPI 점수가 낮다는 것의 의미는?

### ✅ 우리가 정의한 KPI 점수란?
- 여러 지표들을 **표준화(Z-score)** 한 뒤, **평균을 내어 산출**한 종합 점수입니다.
- 포함된 주요 지표:
  - `증감율_이용건수_*` (할부, 일시불, 체크카드 등)
  - `증감율_이용금액_*`
  - `변동률_평잔_*` (할부평잔, 카드론평잔 등)
- **즉**, 최근 이용 건수, 금액, 잔액 등의 **변화율을 종합적으로 반영**한 지표입니다.

---

### 📉 KPI 점수가 낮다는 것은?
- **최근 이용 실적(건수/금액)의 감소 또는 정체**를 의미합니다.
- 지표의 변화율이 **낮거나 음수**인 경우가 많습니다.
- 예시:  
  `증감율_이용금액_할부_전월 = -0.2` → 전월 대비 20% 감소

---

### 🧭 KPI 점수 ↓ = 활동성 ↓?
✅ **맞습니다!** 우리의 KPI 점수는 이용 변화 중심 지표로 구성되어 있기 때문에:

| 해석 요소     | 의미                                       |
|--------------|--------------------------------------------|
| KPI 점수 ↓   | 이용 건수·금액의 변화율이 낮음             |
| 활동성 ↓     | 거래 또는 금융 활동이 줄어들었음            |
| 이탈 가능성 ↑ | 고객 가치 감소 or 이탈 전조로 해석 가능     |

---

### ✅ 결론
- KPI 점수가 낮다는 것은 **고객의 최근 사용량이 줄었거나 변화가 거의 없다는 뜻**입니다.
- 따라서 **고객의 활동성이 낮아졌거나, 이탈 위험이 높다는 지표**로 활용할 수 있습니다.

> 📌 추가 아이디어: KPI 점수를 기준으로 High / Mid / Low 그룹으로 나누어 세그먼트 분석을 확장할 수 있습니다.
