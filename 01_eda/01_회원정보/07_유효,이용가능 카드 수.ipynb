```python
import pandas as pd
import matplotlib.pyplot as plt

# 카드 수 관련 컬럼만 추출
card_cols = [col for col in df.columns if col.startswith('유효카드수') or col.startswith('이용가능카드수') or col.startswith('이용카드수')]

# 세그먼트별 평균값 계산
segment_card_mean = df.groupby('Segment')[card_cols].mean()

# 시각화: 전반적인 카드 수 비교
plt.figure(figsize=(14, 6))
segment_card_mean.T.plot(kind='bar', figsize=(14, 6))
plt.title("세그먼트별 카드 수 평균 비교 (유효/이용가능/이용)")
plt.ylabel("평균 카드 수")
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.tight_layout()
plt.show()
```

- C: 유효카드수 자체는 많음(다양한 카드를 많이 소지) -> 이용가능카드수, 이용카드수 순으로 점점 적어짐 -> A가 역전
    - 보유한 카드를 적극적으로 사용하지 않음 -> 비활성 고객

```python
# 각 세그먼트별 총 유효카드 수와 이용카드 수 합계
card_usage = df.groupby('Segment').agg({
    '유효카드수_신용': 'sum',
    '이용카드수_신용': 'sum',
    '유효카드수_체크': 'sum',
    '이용카드수_체크': 'sum'
})

# 사용률 계산 (0 division 방지)
card_usage['사용률_신용'] = card_usage['이용카드수_신용'] / card_usage['유효카드수_신용'].replace(0, 1)
card_usage['사용률_체크'] = card_usage['이용카드수_체크'] / card_usage['유효카드수_체크'].replace(0, 1)

# 시각화
card_usage[['사용률_신용', '사용률_체크']].plot(kind='bar', figsize=(10,5))
plt.title("세그먼트별 카드 사용률 비교 (신용 / 체크)")
plt.ylabel("카드 사용률")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 각 세그먼트별 총 유효카드 수와 이용카드 수 합계
card_usage = df.groupby('Segment').agg({
    '유효카드수_신용': 'sum',
    '이용가능카드수_신용': 'sum',
    '유효카드수_체크': 'sum',
    '이용가능카드수_체크': 'sum'
})

# 사용률 계산 (0 division 방지)
card_usage['이용가능률_신용'] = card_usage['이용가능카드수_신용'] / card_usage['유효카드수_신용'].replace(0, 1)
card_usage['이용가능률_체크'] = card_usage['이용가능카드수_체크'] / card_usage['유효카드수_체크'].replace(0, 1)

# 시각화
card_usage[['이용가능률_신용', '이용가능률_체크']].plot(kind='bar', figsize=(10,5))
plt.title("세그먼트별 카드 이용가능률 비교 (신용 / 체크)")
plt.ylabel("카드 사용률")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
