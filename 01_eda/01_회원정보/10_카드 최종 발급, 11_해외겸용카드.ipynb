## 11. 해외겸용카드 여부 및 이용

- '보유여부_해외겸용_본인', '이용가능여부_해외겸용_본인', '이용여부_3M_해외겸용_본인', 
- '보유여부_해외겸용_신용_본인', '이용가능여부_해외겸용_신용_본인', '이용여부_3M_해외겸용_신용_본인'

```python
import pandas as pd

# 분석에 필요한 컬럼만 지정
usecols = [ '보유여부_해외겸용_본인', '이용가능여부_해외겸용_본인', '이용여부_3M_해외겸용_본인',
'보유여부_해외겸용_신용_본인', '이용가능여부_해외겸용_신용_본인', '이용여부_3M_해외겸용_신용_본인', 'Segment', 'ID']

# 필요한 컬럼만 불러오기
df = pd.read_csv("../data/train/1.회원정보/train_회원정보.csv", usecols=usecols)
df
```

```python
df.groupby('Segment')[['보유여부_해외겸용_본인', '이용가능여부_해외겸용_본인']].mean().T
```

```python
df.groupby('Segment')[['보유여부_해외겸용_신용_본인', '이용가능여부_해외겸용_신용_본인']].mean().T
```

```python
df.groupby('Segment')[['보유여부_해외겸용_본인', '이용여부_3M_해외겸용_본인']].mean().T
```

```python
df.groupby('Segment')[['보유여부_해외겸용_신용_본인', '이용여부_3M_해외겸용_신용_본인']].mean().T
```

```python
import matplotlib.pyplot as plt

segments = ['A', 'B', 'C', 'D', 'E']
ownership = [0.987654, 0.958333, 0.876142, 0.798283, 0.648592]  # 신용_보유
usage_3m = [0.981481, 0.958333, 0.844690, 0.752667, 0.495193]  # 신용_3M이용

plt.figure(figsize=(8,6))
plt.scatter(ownership, usage_3m)

for i, txt in enumerate(segments):
    plt.annotate(txt, (ownership[i], usage_3m[i]))

plt.xlabel("보유 여부 (신용카드, 본인)")
plt.ylabel("최근 3개월 이용 여부 (신용카드, 본인)")
plt.title("세그먼트별 카드 보유 vs 3개월 이용")
plt.show()
```

```python
df_diff = pd.DataFrame({
    'Segment': segments,
    '차이(3개월이용 - 보유)': [u - o for u, o in zip(usage_3m, ownership)]
})

plt.figure(figsize=(8,5))
sns.barplot(x='Segment', y='차이(3개월이용 - 보유)', data=df_diff)
plt.title("세그먼트별 보유 대비 최근 3개월 이용률 차이")
plt.axhline(0, color='gray', linestyle='--')
plt.ylabel("차이 비율")
plt.show()
```

```python
import matplotlib.pyplot as plt

segments = ['A', 'B', 'C', 'D', 'E']
ownership = [0.987654,	0.958333,	0.934658,	0.891482,	0.751909]  # 보유
usage_3m = [0.987654,	0.958333,	0.887789,	0.820050,	0.547432]  # 신용_3M이용

plt.figure(figsize=(8,6))
plt.scatter(ownership, usage_3m)

for i, txt in enumerate(segments):
    plt.annotate(txt, (ownership[i], usage_3m[i]))

plt.xlabel("보유 여부 (신용카드, 본인)")
plt.ylabel("최근 3개월 이용 여부 (신용카드, 본인)")
plt.title("세그먼트별 카드 보유 vs 3개월 이용")
plt.show()
```

```python
df_diff = pd.DataFrame({
    'Segment': segments,
    '차이(3개월이용 - 보유)': [u - o for u, o in zip(usage_3m, ownership)]
})

plt.figure(figsize=(8,5))
sns.barplot(x='Segment', y='차이(3개월이용 - 보유)', data=df_diff)
plt.title("세그먼트별 보유 대비 최근 3개월 이용률 차이")
plt.axhline(0, color='gray', linestyle='--')
plt.ylabel("차이 비율")
plt.show()
```

```python
import seaborn as sns

data = {
    '보유': ownership,
    '이용_3M': usage_3m
}
df_heatmap = pd.DataFrame(data, index=segments)

plt.figure(figsize=(6,4))
sns.heatmap(df_heatmap, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("세그먼트별 카드 보유 vs 3개월 이용률")
plt.ylabel("Segment")
plt.show()
```
