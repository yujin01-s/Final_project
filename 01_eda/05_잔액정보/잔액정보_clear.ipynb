```python
import pandas as pd

# 1. Segment 정보 불러오기
member_df = pd.read_csv("../data/train/1.회원정보/train_회원정보.csv", usecols=["ID", "Segment"])

# 2. 잔액정보 불러오기
balance_df = pd.read_csv("../data/train/5.잔액정보/train_잔액정보.csv")

# 3. 병합
merged_df = pd.merge(balance_df, member_df, on="ID", how="left")

# 4. 결과 저장
merged_df.to_csv("../data/train/5.잔액정보/train_잔액정보_with_segment.csv", index=False)

print("✅ 잔액정보 + Segment 병합 완료! 저장 경로: train_잔액정보_with_segment.csv")
```

```python
import pandas as pd
df = pd.read_csv("../data/train/5.잔액정보/train_잔액정보_with_segment.csv")
df
```

```python
# 기준년월과 ID 조합이 중복된 행 중 1개만 남김
df_unique = df.drop_duplicates(subset=['기준년월', 'ID']).copy()
```

```python
# 중복 제거 후에도 여전히 중복되는 행이 있는지 확인
dup_count = df_unique.duplicated(subset=['기준년월', 'ID']).sum()
print(f"중복된 (기준년월, ID) 조합 수: {dup_count}")  # 0이면 성공
```

```python
# 각 ID가 몇 개의 기준년월 데이터를 가지고 있는지 확인
month_counts = df_unique.groupby('ID')['기준년월'].nunique()
print(month_counts.value_counts().sort_index())
```
