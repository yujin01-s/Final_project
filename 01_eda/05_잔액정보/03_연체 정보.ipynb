```python
import pandas as pd

# 분석에 필요한 컬럼만 지정
cols = ['연체일자_B0M', '연체잔액_B0M', '연체잔액_일시불_B0M', '연체잔액_할부_B0M', '연체잔액_현금서비스_B0M', 
        '연체잔액_카드론_B0M', '연체잔액_대환론_B0M','연체잔액_CA_B0M',
        '연체일수_B1M', '연체일수_B2M', '연체원금_B1M', '연체원금_B2M',
        '연체일수_최근', '연체원금_최근', '최종연체회차', '최종연체개월수_R15M', '기준년월', 'Segment', 'ID']

# 필요한 컬럼만 불러오기
df = pd.read_csv("../data/train/5.잔액정보/train_잔액정보_clesr.csv", usecols=cols)
df
```

```python
print("데이터 타입 확인:")
print(df.dtypes)

print("\n결측치 개수 확인:")
print(df.isnull().sum())
```

```python
df['연체잔액_카드론_B0M'].value_counts()
```

```python
zero_df = df[df['연체잔액_카드론_B0M'] == 0]
zero_segment_pct = zero_df['Segment'].value_counts(normalize=True).sort_index() * 100
print("연체잔액_카드론_B0M == 0 인 경우의 Segment 비율 (%):\n", zero_segment_pct)
```

```python
zero_df = df[df['연체잔액_B0M'] != 0]
zero_segment_pct = zero_df['Segment'].value_counts(normalize=True).sort_index() * 100
print("연체잔액_B0M != 0 인 경우의 Segment 비율 (%):\n", zero_segment_pct)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 분석 대상 컬럼
columns = [
    '연체잔액_B0M',
    '연체잔액_일시불_B0M',
    '연체잔액_할부_B0M',
    '연체잔액_현금서비스_B0M',
    '연체잔액_CA_B0M'
]

# 평균값 계산
avg_df = df.groupby('Segment')[columns].mean().reset_index()

# 시각화
plt.figure(figsize=(12, 6))
avg_df_melted = avg_df.melt(id_vars='Segment', var_name='연체유형', value_name='평균연체잔액')

sns.barplot(data=avg_df_melted, x='연체유형', y='평균연체잔액', hue='Segment')
plt.title("Segment별 연체 유형별 평균 연체잔액")
plt.ylabel("평균 연체 잔액 (원)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Segment A, B만 필터링
ab_df = avg_df_melted[avg_df_melted['Segment'].isin(['A', 'B'])]

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=ab_df, x='연체유형', y='평균연체잔액', hue='Segment')
plt.title("Segment A, B 연체 유형별 평균 연체잔액")
plt.ylabel("평균 연체 잔액 (원)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
