## 1. 전체 컬럼 목록 및 결측치 비율 확인

```python
import pandas as pd

# 병합된 승인매출정보 데이터 불러오기
df = pd.read_parquet("merged_승인매출정보.parquet")

# 전체 컬럼 수 및 컬럼 리스트
print("전체 컬럼 수:", len(df.columns))
print("컬럼 목록:")
for col in df.columns:
    print("-", col)
```

```python
df.head(30)
```

```python
# 결측치 비율 계산
missing_ratio = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
missing_df = pd.DataFrame({
    '컬럼명': missing_ratio.index,
    '결측률(%)': missing_ratio.round(2).values
})

# 상위 20개만 보기
missing_df.head(30)
```

## 🔍 업종 관련 컬럼들의 결측치 분석

- 업종이라는 키워드가 포함된 컬럼만 따로 필터링하여 결측률을 분석합니다.
- 결측치를 직접 채우지 않고, **결측률이 낮은 컬럼만 모델링 대상 후보로 선정**합니다.
- 결측률이 높은 컬럼은 분석 시 제외할 수 있습니다.

```python
# '업종'이라는 단어가 컬럼명에 포함된 컬럼만 필터링
업종_컬럼 = [col for col in df.columns if '업종' in col]

# 해당 컬럼들의 결측률 계산
업종_결측률 = df[업종_컬럼].isnull().mean().sort_values(ascending=False) * 100

# 데이터프레임으로 보기 좋게 정리
업종_결측률_df = pd.DataFrame({
    '컬럼명': 업종_결측률.index,
    '결측률(%)': 업종_결측률.values.round(2)
}).reset_index(drop=True)

# 결과 확인
업종_결측률_df
```

## 📊 업종별 segment 분포 비교

업종 컬럼을 기준으로 고객들이 어떤 segment에 많이 속해 있는지를 확인합니다.

이 분석을 통해 특정 업종과 segment 간의 연관성(상관관계, 분포 차이 등)을 가설로 제시할 수 있습니다.

### _1순위업종 기준 분포 분석

```python
import pandas as pd
import glob

# 모든 승인매출 parquet 파일 경로 가져오기
file_paths = glob.glob("../data/train/1.회원정보/*_train_회원정보.parquet")

# 병합
id_df = pd.concat([pd.read_parquet(fp) for fp in sorted(file_paths)], ignore_index=True)

# 병합 결과 저장
id_df.to_parquet("merged_회원정보.parquet", index=False)
```

```python
# df에서 필요한 컬럼만 유지
df = df[['_1순위업종', 'ID']]  # 또는 ID 컬럼명에 맞게 조정

# 회원정보도 필요한 컬럼만 유지
id_df = id_df[['ID', 'Segment']]  # 정확한 컬럼명으로 조정

# 병합 수행 (줄어든 상태로)
df_merged = df.merge(id_df, on='ID', how='left')
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 윈도우 기준: 기본 시스템 폰트 중 하나 사용
plt.rcParams['font.family'] = 'Pretendard'  # Windows
plt.rcParams['axes.unicode_minus'] = False     # 마이너스 부호 깨짐 방지
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 업종별 segment 분포 비율표 생성 (normalize='index' → 비율 기준)
cross_tab = pd.crosstab(df_merged['_1순위업종'], df_merged['Segment'], normalize='index') * 100
cross_tab = cross_tab.round(1).sort_values(by='E', ascending=False)

# 상위 10개 업종만 시각화
top_e = cross_tab['E'].head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_e.values, y=top_e.index, palette='Blues_d')
plt.title("Segment E 비율이 높은 업종 TOP 10")
plt.xlabel("Segment E 비율 (%)")
plt.ylabel("_1순위업종")
plt.tight_layout()
plt.show()
```

```python
# 업종별 segment 분포 비율표 생성
cross_tab = pd.crosstab(df_merged['_1순위업종'], df_merged['Segment'], normalize='index') * 100
cross_tab = cross_tab.round(2)

# Segment A 기준 내림차순
top_a = cross_tab.sort_values(by='A', ascending=False).head(10)

# Segment B 기준 내림차순
top_b = cross_tab.sort_values(by='B', ascending=False).head(10)

# 결과 확인
print("🔹 Segment A 비율이 높은 업종 TOP 10")
display(top_a)

print("🔹 Segment B 비율이 높은 업종 TOP 10")
display(top_b)
```

- 전체적으로 Segment A/B 비율은 매우 희박하며, 대부분의 업종에서 Segment E 비율이 절대적으로 높습니다.
- 그러나 **의료, 쇼핑, 사교활동, 교육**과 같은 업종에서는 Segment A 비율이 상대적으로 높게 나타나,  
  희귀 세그먼트 분류에 있어 중요한 구분 변수로 작용할 가능성이 있습니다.

## 📊 _2순위업종과 Segment A/B 분포 비교

앞서 `_1순위업종`을 기준으로 Segment A와 B에 특화된 업종을 확인했습니다.  
이번에는 `_2순위업종`을 기준으로 동일한 방식으로 분석을 진행하여,  
희귀 클래스에 유의미한 업종이 존재하는지 파악합니다.

```python
import pandas as pd

# 필요한 컬럼만 불러오기 (업종 컬럼 + ID 정도만)
columns = ['ID', '_1순위업종', '_2순위업종', '_3순위업종']
df = pd.read_parquet("merged_승인매출정보.parquet", columns=columns)


# 회원정보에서 필요한 컬럼만 추출
회원정보 = pd.read_parquet("merged_회원정보.parquet")

# 컬럼명 확인 후 정확하게 선택
회원정보 = 회원정보[['ID', 'Segment']]
# 병합 (ID 기준)
df_merged = df.merge(회원정보, on='ID', how='left')
```

```python
# Segment 비율 계산 (_2순위업종 기준)
cross_tab_2 = pd.crosstab(df_merged['_2순위업종'], df_merged['Segment'], normalize='index') * 100
cross_tab_2 = cross_tab_2.round(2)

# Segment A 기준 상위 10개
top_a_2 = cross_tab_2.sort_values(by='A', ascending=False).head(10)

# Segment B 기준 상위 10개
top_b_2 = cross_tab_2.sort_values(by='B', ascending=False).head(10)

# 결과 확인
print("🔹 Segment A 비율 높은 _2순위업종")
display(top_a_2)

print("🔹 Segment B 비율 높은 _2순위업종")
display(top_b_2)
```

```python
# Segment 비율 계산 (_3순위업종 기준)
cross_tab_3 = pd.crosstab(df_merged['_3순위업종'], df_merged['Segment'], normalize='index') * 100
cross_tab_3 = cross_tab_3.round(2)

# Segment A 기준 상위 10개
top_a_3 = cross_tab_3.sort_values(by='A', ascending=False).head(10)

# Segment B 기준 상위 10개
top_b_3 = cross_tab_3.sort_values(by='B', ascending=False).head(10)

# 결과 확인
print("🔹 Segment A 비율 높은 _3순위업종")
display(top_a_3)

print("🔹 Segment B 비율 높은 _3순위업종")
display(top_b_3)
```

"_3순위업종" 분석 결과, 특히 **'교육' 업종**은 Segment A/B 비율이 눈에 띄게 높은 업종으로 확인되었고,  
Segment E 비중이 낮기 때문에 **희귀 클래스 분류 모델링에 있어서 구분력이 높은 변수**가 될 수 있음을 시사한다.

또한 '의료', '사교활동', '해외' 등의 업종은 `_1~3순위` 모두에서 반복적으로 등장하며,  
Segment A/B 식별 변수로 일관된 패턴을 보였다.

| 업종      | `_1순위` A | `_2순위` A | `_3순위` A | 판단              |
| ------- | -------- | -------- | -------- | --------------- |
| 교육      | 0.03     | 0.35     | **0.68** | ⏫ 일관된 A 유의미 업종  |
| 의료      | 0.20     | 0.30     | 0.25     | ⏫ 세 순위 모두 등장    |
| 사교활동    | 0.05     | 0.08     | 0.08     | ⏫ 고르게 약한 영향     |
| 쇼핑      | 0.08     | 0.03     | 0.02     | ⬇ 순위 낮을수록 비중 감소 |
| 요식/일상생활 | 0.00     | 0.00     | 0.01     | ⬇ A/B 분류 도움 X   |
