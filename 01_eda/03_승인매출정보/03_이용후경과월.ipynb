```python
import pandas as pd

# 병합된 승인매출정보 데이터 불러오기
df = pd.read_parquet("merged_승인매출정보.parquet")

# 전체 컬럼 수 및 컬럼 리스트
print("전체 컬럼 수:", len(df.columns))
print("컬럼 목록:")
for col in df.columns:
    print("-", col)
```

```python
# 회원정보에서 필요한 컬럼만 추출
df_member = pd.read_parquet("merged_회원정보.parquet")
```

```python
# 이용후경과월 컬럼만 추출
elapsed_cols = [col for col in df.columns if "이용후경과월_" in col]
elapsed_cols = df[['ID'] + elapsed_cols]  # ID 추가해서 서브셋 생성

# 2. 회원정보에서 ID와 Segment만 추출
df_member_subset = df_member[['ID', 'Segment']]

# 3. 병합
merged_df = elapsed_cols.merge(df_member_subset, on='ID', how='left')
merged_df
```

```python
# 결측치 확인
print(merged_df.isnull().sum())
```

```python
# 1. 세그먼트별 평균 이용후경과월 계산
segment_avg = merged_df.groupby('Segment')[
    ['이용후경과월_신용', '이용후경과월_신판', '이용후경과월_일시불',
     '이용후경과월_할부', '이용후경과월_할부_유이자', '이용후경과월_할부_무이자',
     '이용후경과월_부분무이자', '이용후경과월_CA', '이용후경과월_체크', '이용후경과월_카드론']
].mean().round(2)

# 2. 출력
print("세그먼트별 평균 이용후경과월")
display(segment_avg)
```
