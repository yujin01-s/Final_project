### 이용건수 vs 이용금액 상관관계

```python
# '페이'가 컬럼명에 포함된 컬럼 필터링
pay_col = [col for col in merged_df.columns if '페이' in col]

# 해당 컬럼 제거
merged_df = merged_df.drop(columns=pay_col)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 이용건수 및 이용금액 컬럼만 추출
count_cols = [col for col in df.columns if col.startswith('이용건수') and '_B0M' in col]
amount_cols = [col for col in df.columns if col.startswith('이용금액') and '_B0M' in col]

# count vs amount correlation matrix
correlation_matrix = df[count_cols + amount_cols].corr().loc[count_cols, amount_cols]

# 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("이용건수 vs 이용금액 상관관계 히트맵")
plt.xlabel("이용금액 항목")
plt.ylabel("이용건수 항목")
plt.tight_layout()
plt.show()
```

```python
# 조건 1: 'RP' 또는 '페이'가 포함되지 않은 컬럼
condition1 = ~segment_mean.columns.str.contains('RP|페이')

# 조건 2: '이용건수' 또는 '이용금액'이 포함된 컬럼
condition2 = segment_mean.columns.str.contains('이용건수|이용금액')

# 두 조건을 모두 만족하는 컬럼만 선택
valid_cols = segment_mean.columns[condition1 & condition2]
filtered_segment_mean = segment_mean[valid_cols]
```

```python
# 이용건수 / 이용금액 변수 분리
usage_count_cols = [col for col in filtered_segment_mean.columns if '이용건수' in col]
usage_amount_cols = [col for col in filtered_segment_mean.columns if '이용금액' in col]

# 히트맵 1: 이용건수
plt.figure(figsize=(14, 6))
sns.heatmap(filtered_segment_mean[usage_count_cols].T, cmap="YlGnBu", annot=False)
plt.title("Segment별 이용건수 평균값 히트맵")
plt.xlabel("Segment")
plt.ylabel("이용건수 변수")
plt.tight_layout()
plt.show()

# 히트맵 2: 이용금액
plt.figure(figsize=(14, 6))
sns.heatmap(filtered_segment_mean[usage_amount_cols].T, cmap="YlGnBu", annot=False)
plt.title("Segment별 이용금액 평균값 히트맵")
plt.xlabel("Segment")
plt.ylabel("이용금액 변수")
plt.tight_layout()
plt.show()
```
