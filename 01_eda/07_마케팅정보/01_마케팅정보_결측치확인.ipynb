```python
member_df = pd.read_csv('open/train/2018_train_회원정보.csv')
member_df
```

```python
# 1. 회원정보에서 Segment 컬럼만 추출 + 중복 제거
segment_df = member_df[['ID', 'Segment']].drop_duplicates()
segment_df
```
