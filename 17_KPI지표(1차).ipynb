```python
from sklearn.preprocessing import MinMaxScaler

# ÏÇ¨Ïö©Ìï† KPI Ïª¨Îüº Î™©Î°ù
kpi_cols = [
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ìåê_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ìï†Î∂Ä_Ï†ÑÏõî',
    'Î≥ÄÎèôÎ•†_RVCAÌèâÏûî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_Î∂ÑÍ∏∞',
    'Î≥ÄÎèôÎ•†_Ïπ¥ÎìúÎ°†ÌèâÏûî'
]

# Í≤∞Ï∏°Ïπò Ï†úÍ±∞ (ÌòπÏãú Î™®Î•¥Îãà ÌïúÎ≤à Îçî ÌôïÏù∏)
result_df_kpi = result_df.dropna(subset=kpi_cols)

# ÌëúÏ§ÄÌôî (0~1 Ïä§ÏºÄÏùºÎßÅ)
scaler = MinMaxScaler()
result_df_kpi_scaled = result_df_kpi.copy()
result_df_kpi_scaled[kpi_cols] = scaler.fit_transform(result_df_kpi_scaled[kpi_cols])

# KPI Ïä§ÏΩîÏñ¥ Í≥ÑÏÇ∞ (Í∞ÄÏ§ëÏπò ÏóÜÏù¥ Îã®Ïàú ÌèâÍ∑†)
result_df_kpi_scaled['KPI_SCORE'] = result_df_kpi_scaled[kpi_cols].mean(axis=1)
```

### Í∞ÄÏ§ëÏπòÎ•º Îã§Ïãú Îëò Ïàò ÏûàÎäî Î∞©Î≤ï Ïû¨Í≥†
- ÏÑ∏Í∑∏Î®ºÌä∏ Í∞Ñ ÌèâÍ∑†Ïùò Ï∞®Ïù¥Î•º Í∏∞Ï§ÄÏúºÎ°ú KPI Ïª¨ÎüºÎ≥Ñ Í∞ÄÏ§ëÏπòÎ•º Î∂ÄÏó¨ÌïòÎäî Î∞©Î≤ï

```python
import pandas as pd

# KPI ÌõÑÎ≥¥ Ïª¨ÎüºÎì§ (ÏòàÏãú Î¶¨Ïä§Ìä∏Î°ú ÏûëÏÑ±Ìï¥Îë°ÎãàÎã§)
kpi_cols = [
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ìï†Î∂Ä_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_CA_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ïö©_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ìåê_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_ÏùºÏãúÎ∂à_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ìï†Î∂Ä_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_Î∂ÑÍ∏∞',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_Î∂ÑÍ∏∞',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ïö©_Î∂ÑÍ∏∞',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_Î∂ÑÍ∏∞',
    'Î≥ÄÎèôÎ•†_RVÏùºÏãúÎ∂àÌèâÏûî',
    'Î≥ÄÎèôÎ•†_Ìï†Î∂ÄÌèâÏûî',
    'Î≥ÄÎèôÎ•†_RVCAÌèâÏûî',
    'Î≥ÄÎèôÎ•†_Ïπ¥ÎìúÎ°†ÌèâÏûî'
]

# 1. Segment Î≥Ñ ÌèâÍ∑† Íµ¨ÌïòÍ∏∞
segment_means = result_df.groupby("Segment")[kpi_cols].mean()

# 2. Í∞Å Ïª¨ÎüºÎ≥Ñ (ÏµúÎåÄ - ÏµúÏÜå) Î≤îÏúÑ Í≥ÑÏÇ∞
mean_range = segment_means.max() - segment_means.min()

# 3. Ï†ÑÏ≤¥ Ï¥ùÌï©ÏúºÎ°ú ÎÇòÎà†ÏÑú Í∞ÄÏ§ëÏπò Í≥ÑÏÇ∞
weights = mean_range / mean_range.sum()

# 4. Í∞Å rowÏóê Í∞ÄÏ§ëÏπòÎ•º Í≥±Ìï¥ÏÑú KPI Score Í≥ÑÏÇ∞
result_df['KPI_Score'] = result_df[kpi_cols].mul(weights, axis=1).sum(axis=1)

# Í≤∞Í≥º ÌôïÏù∏ (ÏÉÅÏúÑ Î™á Í∞úÎßå Ï∂úÎ†•)
print(result_df[['Segment', 'KPI_Score']].head())
```

- ÌòπÏùÄ KPI Í≥ÑÏÇ∞Ïóê ÏÇ¨Ïö©Ìïú Ïª¨ÎüºÎ≥ÑÎ°ú **Í∞Å ÏÑ∏Í∑∏Î®ºÌä∏Ïùò ÌèâÍ∑†Í∞í ÎπÑÍµê**Î•º Îã§Ïãú Ìï¥Î≥¥Î©¥ Ïñ¥Îñ§ Ìï≠Î™©ÏóêÏÑú A, BÍ∞Ä ÎÇÆÏùÄÏßÄ ÌååÏïÖ Í∞ÄÎä•.


### Í∞Å ÏÑ∏Í∑∏Î®ºÌä∏ Î≥ÑÎ°ú ÎèôÏùºÌïú ÏÉòÌîå ÏàòÎ•º Ï∂îÏ∂ú

```python
import pandas as pd

# ÏÑ∏Í∑∏Î®ºÌä∏Î≥Ñ ÏµúÏÜå ÏÉòÌîå Ïàò ÌôïÏù∏
min_count = result_df['Segment'].value_counts().min()
print(f"ÏµúÏÜå ÏÉòÌîå Ïàò: {min_count}")

# Í∞Å ÏÑ∏Í∑∏Î®ºÌä∏Î≥ÑÎ°ú ÎèôÏùºÌïòÍ≤å ÏÉòÌîåÎßÅ
balanced_df = (
    result_df.groupby('Segment')
    .apply(lambda x: x.sample(n=min_count, random_state=42))
    .reset_index(drop=True)
)
```

```python
# KPI Ï†êÏàò Í≥ÑÏÇ∞
from sklearn.preprocessing import StandardScaler

# KPI ÌõÑÎ≥¥ Ïª¨Îüº Î¶¨Ïä§Ìä∏
kpi_cols = [
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ìï†Î∂Ä_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_CA_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ï≤¥ÌÅ¨_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ïö©_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ìåê_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_ÏùºÏãúÎ∂à_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ìï†Î∂Ä_Ï†ÑÏõî',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_Ïã†Ìåê_Î∂ÑÍ∏∞',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í±¥Ïàò_ÏùºÏãúÎ∂à_Î∂ÑÍ∏∞',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ïã†Ïö©_Î∂ÑÍ∏∞',
    'Ï¶ùÍ∞êÏú®_Ïù¥Ïö©Í∏àÏï°_Ï≤¥ÌÅ¨_Î∂ÑÍ∏∞',
    'Î≥ÄÎèôÎ•†_RVÏùºÏãúÎ∂àÌèâÏûî',
    'Î≥ÄÎèôÎ•†_Ìï†Î∂ÄÌèâÏûî',
    'Î≥ÄÎèôÎ•†_RVCAÌèâÏûî',
    'Î≥ÄÎèôÎ•†_Ïπ¥ÎìúÎ°†ÌèâÏûî'
]

# Í≤∞Ï∏°Ïπò Ï†úÍ±∞ (ÌòπÏãú ÏûàÏùÑ Í≤ΩÏö∞)
kpi_data = balanced_df[kpi_cols].dropna()

# Z-score ÌëúÏ§ÄÌôî
scaler = StandardScaler()
kpi_scaled = scaler.fit_transform(kpi_data)

# KPI Score Í≥ÑÏÇ∞: ÌëúÏ§ÄÌôîÎêú ÏßÄÌëúÎì§Ïùò ÌèâÍ∑†
balanced_df = balanced_df.loc[kpi_data.index].copy()
balanced_df['KPI_Score'] = kpi_scaled.mean(axis=1)
```

## ‚úÖ Í∑†Îì± ÏÉòÌîåÎßÅ Í∏∞Î∞ò ÏÑ∏Í∑∏Î®ºÌä∏Î≥Ñ KPI Score Î∂ÑÌè¨ Ìï¥ÏÑù

1. **ÏÉòÌîå Î∂àÍ∑†Ìòï Î≥¥Ï†ï ÏÑ±Í≥µ**
   - Ïù¥Ï†ÑÏóêÎäî A, B ÏÑ∏Í∑∏Î®ºÌä∏Í∞Ä ÏÉòÌîå ÏàòÍ∞Ä Ï†ÅÏñ¥ÏÑú ÌèâÍ∑† Í∏∞Î∞ò Î∂ÑÏÑùÏóêÏÑú Ï†ÄÌèâÍ∞ÄÎêòÏóàÏßÄÎßå,
   - Í∑†Îì± ÏÉòÌîåÎßÅ ÌõÑ Í∞Å ÏÑ∏Í∑∏Î®ºÌä∏Í∞Ä ÎèôÏùºÌïú Ï°∞Í±¥ÏóêÏÑú ÎπÑÍµêÎêòÎ©∞ Îçî Í≥µÏ†ïÌïú Î∂ÑÏÑùÏù¥ Í∞ÄÎä•Ìï¥Ï°åÏùå.

2. **ÏÑ∏Í∑∏Î®ºÌä∏Î≥Ñ Î∂ÑÌè¨ ÌäπÏßï**
   - **D, E** ÏÑ∏Í∑∏Î®ºÌä∏Îäî ÌèâÍ∑† KPI ScoreÍ∞Ä ÏÉÅÎåÄÏ†ÅÏúºÎ°ú ÎÜíÍ≥†, **ÏÉÅÏúÑ Ïù¥ÏÉÅÏπò(Outlier)**ÎèÑ ÏûêÏ£º Î∞úÏÉùÌï®.
   - Î∞òÎ©¥ **B**Îäî Ï§ëÏïôÍ∞íÏù¥ 0Î≥¥Îã§ ÎÇÆÏúºÎ©∞ Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú ÏÑ±Í≥º ÏßÄÌëúÍ∞Ä ÎÇÆÏùÄ Î∂ÑÌè¨Î•º Î≥¥ÏûÑ.
   - **A, C**Îäî Ï§ëÍ∞Ñ ÏàòÏ§ÄÏùò ÏïàÏ†ïÏ†Å ÏÑ±Í≥º Î∂ÑÌè¨Î°ú Ìï¥ÏÑù Í∞ÄÎä•.

3. **ÌôúÎèôÏÑ±Í≥º KPIÏùò ÎπÑÏÑ†Ìòï Í¥ÄÍ≥Ñ Í∞ÄÎä•ÏÑ±**
   - ÌôúÎèôÏÑ±Ïù¥ ÎÜíÏïòÎçò A, B ÏÑ∏Í∑∏Î®ºÌä∏Í∞Ä KPI ScoreÏóêÏÑúÎäî ÏÉÅÏúÑÍ∞Ä ÏïÑÎãê Ïàò ÏûàÎã§Îäî Ï†êÏùÄ,
   - Îã®Ïàú ÌôúÎèôÏÑ±Î≥¥Îã§ **Î≥ÄÎèôÏÑ±, Ï¶ùÍ∞ÄÏú®, Î¶¨Ïä§ÌÅ¨ ÏöîÏù∏** Îì±Ïù¥ KPIÏóê Îçî ÌÅ∞ ÏòÅÌñ•ÏùÑ Ï£ºÏóàÏùåÏùÑ ÏãúÏÇ¨Ìï®.

4. **ÏãúÏÇ¨Ï†ê**
   - KPIÎäî Îã®Ïàú ÌèâÍ∑†Î≥¥Îã§Îäî ÌëúÏ§ÄÌôîÏôÄ ÏÉòÌîåÎßÅÏùÑ Í≥†Î†§Ìï¥ Î∂ÑÏÑùÌï† ÌïÑÏöîÍ∞Ä ÏûàÏùå.
   - Ïù¥ Í≤∞Í≥ºÎäî Ïù¥ÌõÑ ÏÑ∏Í∑∏Î®ºÌä∏ Î¶¨Ìè¨Ìä∏ ÎòêÎäî ÎßàÏºÄÌåÖ Ï†ÑÎûµ ÏàòÎ¶ΩÏóê Í∏∞Ï¥à ÏßÄÌëúÎ°ú ÏÇ¨Ïö©Îê† Ïàò ÏûàÏùå.


```python
import scipy.stats as stats

# ÏÑ∏Í∑∏Î®ºÌä∏Î≥Ñ KPI Score Í∑∏Î£π ÎÇòÎàÑÍ∏∞
groups = [group["KPI_Score"].values for name, group in result_df.groupby("Segment")]

# ÏùºÏõêÎ∂ÑÏÇ∞Î∂ÑÏÑù (One-Way ANOVA)
f_stat, p_val = stats.f_oneway(*groups)

print(f"F-statistic: {f_stat:.4f}")
print(f"p-value: {p_val:.4f}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(endog=result_df['KPI_Score'],
                          groups=result_df['Segment'],
                          alpha=0.05)

print(tukey)
```
