```python
bill_df_with_segment['í• ì¸ê±´ìˆ˜_R3M'].value_counts()
```

```python
discount_r3m = bill_df_with_segment.groupby("Segment")['í• ì¸ê±´ìˆ˜_R3M'].value_counts().unstack().fillna(0)

# ì›í•˜ëŠ” ìˆœì„œ ëª…ì‹œì  ì§€ì •
desired_order = ["1íšŒ ì´ìƒ", "10íšŒ ì´ìƒ", "20íšŒ ì´ìƒ", "30íšŒ ì´ìƒ", "40íšŒ ì´ìƒ"]

# discount_r3m êµì°¨í‘œì—ì„œ í•´ë‹¹ ì—´ë§Œ ì¶”ì¶œ
discount_r3m_filtered = discount_r3m[desired_order]
display(discount_r3m_filtered)
```

```python
# í• ì¸ê±´ìˆ˜_R3M ë¹„ìœ¨í‘œ ìƒì„±
discount_r3m_ratio = bill_df_with_segment.groupby("Segment")['í• ì¸ê±´ìˆ˜_R3M'].value_counts(normalize=True).unstack().fillna(0)

# ì›í•˜ëŠ” ìˆœì„œ ëª…ì‹œì ìœ¼ë¡œ ì§€ì •
desired_order = ["1íšŒ ì´ìƒ", "10íšŒ ì´ìƒ", "20íšŒ ì´ìƒ", "30íšŒ ì´ìƒ", "40íšŒ ì´ìƒ"]

# discount_r3m_ratioì—ì„œ í•´ë‹¹ ì—´ë§Œ ì¶”ì¶œ
discount_r3m_ratio_filtered = discount_r3m_ratio[desired_order]

# ì¶œë ¥
from IPython.display import display
display(discount_r3m_ratio_filtered)
```

```python
# 1. ê°€ì¤‘ì¹˜ ì§€ì •
weights = {
    "1íšŒ ì´ìƒ": 1,
    "10íšŒ ì´ìƒ": 10,
    "20íšŒ ì´ìƒ": 20,
    "30íšŒ ì´ìƒ": 30,
    "40íšŒ ì´ìƒ": 40
}

# 2. ê°€ì¤‘ì¹˜ ê³±í•´ì„œ ì ìˆ˜ ê³„ì‚°
discount_score = discount_r3m_ratio_filtered.copy()
for col in discount_score.columns:
    discount_score[col] *= weights[col]

# 3. ì„¸ê·¸ë¨¼íŠ¸ë³„ ì´ ì ìˆ˜ ê³„ì‚°
segment_discount_score = discount_score.sum(axis=1).sort_values(ascending=False)

# 4. ê²°ê³¼ ì¶œë ¥
print("ì„¸ê·¸ë¨¼íŠ¸ë³„ ê°€ì¤‘ ì ìˆ˜ í•©ê³„ (ë†’ì„ìˆ˜ë¡ ìµœê·¼ 3ê°œì›” ë‚´ í• ì¸ê±´ìˆ˜ ì´ìš©ë„ê°€ ë†’ì€ ì„¸ê·¸ë¨¼íŠ¸)")
display(segment_discount_score)
```

```python
# 1. í• ì¸ê±´ìˆ˜_B0Mì´ '10íšŒ ì´ìƒ'ì¸ ê²½ìš°ë§Œ í•„í„°ë§
filtered_df = bill_df_with_segment[bill_df_with_segment['í• ì¸ê±´ìˆ˜_B0M'] == '10íšŒ ì´ìƒ']

# 2. ì„¸ê·¸ë¨¼íŠ¸ë³„ ë„ìˆ˜ í™•ì¸
segment_distribution = filtered_df['Segment'].value_counts().sort_values(ascending=False)

# 3. ì„¸ê·¸ë¨¼íŠ¸ë³„ ë¹„ìœ¨ í™•ì¸
segment_ratio = filtered_df['Segment'].value_counts(normalize=True).sort_values(ascending=False)

# ê²°ê³¼ ì¶œë ¥
from IPython.display import display
print("âœ… í• ì¸ê±´ìˆ˜_B0M = '10íšŒ ì´ìƒ' ì¡°ê±´ì¼ ë•Œ ì„¸ê·¸ë¨¼íŠ¸ë³„ ë¶„í¬:")
display(segment_distribution)

print("\nğŸ“Š ì„¸ê·¸ë¨¼íŠ¸ë³„ ë¹„ìœ¨:")
display(segment_ratio)
```

```python
# ê¸°ë³¸
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ê²½ê³  ëœ¨ì§€ ì•Šê²Œ ì„¤ì •
import warnings
warnings.filterwarnings('ignore')

# ê·¸ë˜í”„ ì„¤ì •
sns.set()

# ê·¸ë˜í”„ ê¸°ë³¸ ì„¤ì •
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False


# ë¶„ì„ì— ì‚¬ìš©í•  ì»¬ëŸ¼ ëª©ë¡
target_columns = [
    'í• ì¸ê¸ˆì•¡_R3M', 
    'í• ì¸ê¸ˆì•¡_B0M', 
    'í• ì¸ê¸ˆì•¡_ì²­êµ¬ì„œ_R3M', 
    'í• ì¸ê¸ˆì•¡_ì²­êµ¬ì„œ_B0M'
]

# ì„¸ê·¸ë¨¼íŠ¸ ë”ë¯¸ ì¸ì½”ë”©
df_corr = bill_df_with_segment.copy()
segment_dummies = pd.get_dummies(df_corr['Segment'], prefix='Segment')
df_corr = pd.concat([df_corr[target_columns], segment_dummies], axis=1)

# ìƒê´€ê³„ìˆ˜ ê³„ì‚°
corr_matrix = df_corr.corr()

# í• ì¸ê¸ˆì•¡ ê´€ë ¨ ì»¬ëŸ¼ê³¼ ì„¸ê·¸ë¨¼íŠ¸ ë”ë¯¸ë§Œ ì¶”ì¶œ
selected_corr = corr_matrix.loc[target_columns, segment_dummies.columns]

# íˆíŠ¸ë§µ ì‹œê°í™”
plt.figure(figsize=(10, 5))
sns.heatmap(selected_corr, annot=True, fmt=".2f", cmap="coolwarm", vmax=1, vmin=-1)
plt.title("í• ì¸ê¸ˆì•¡ ê´€ë ¨ ì»¬ëŸ¼ê³¼ ì„¸ê·¸ë¨¼íŠ¸ ê°„ì˜ ìƒê´€ê´€ê³„")
plt.show()
```
