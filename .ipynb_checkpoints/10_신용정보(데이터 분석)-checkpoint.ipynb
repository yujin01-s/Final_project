## 📌 분석 변수: 월상환론상향가능한도금액

### 🔍 1. Segment별 평균 및 중앙값

- 평균값 기준: Segment B는 **0.75**로 다른 그룹(A, D 등)의 2.8에 비해 현저히 낮음
- 중앙값은 모든 세그먼트에서 **0** → 대부분 고객이 이 변수 값이 없음
- → **분포가 극단적으로 비대칭**, 평균은 일부 극단값에 의해 왜곡된 상태

---

### 📊 2. 분포 시각화 해석

- **히스토그램**: 거의 모든 고객이 0에 밀집, 일부만 값 보유 (long tail 분포)
- **박스플롯**: 0 이상인 값은 극소수 이상치로 나타남
- → 시각적으로도 Segment B는 뚜렷하게 값이 거의 없음

---

### ✅ 3. 0 초과 고객 비율 (%)

| Segment | 0 초과 고객 비율 (%) |
|---------|-----------------------|
| A       | 15.6%                |
| B       | **4.2%**             |
| C       | 12.0%                |
| D       | 15.9%                |
| E       | 15.6%                |

- Segment B는 **0이 아닌 고객 비율이 가장 낮음**
- → **Segment B를 구분하기 위한 강력한 이진 특성 변수**로 활용 가능
- → 연속형보다는 **이진형 변수 변환 후 분석/모델링**에 적합


```python
# 월상환론상향가능한도금액 > 0 여부로 이진 변수 생성
df10['월상환론상향가능한도금액_binary'] = (df10['월상환론상향가능한도금액'] > 0).astype(int)

# 이진 변수 확인
print(df10['월상환론상향가능한도금액_binary'].value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 예측 대상과 특성 선택
features = ['월상환론상향가능한도금액_binary']
X = df10[features]
y = df10['Segment']

# 2. 레이블 인코딩
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 3. 학습/검증 분리
X_train, X_valid, y_train, y_valid = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# 4. 모델 학습
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# 5. 변수 중요도 확인
importances = pd.Series(model.feature_importances_, index=features)
importances.sort_values().plot(kind='barh')
plt.title("🎯 Feature Importance (단일 변수 실험)")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 예측 대상 설정
target = 'Segment'

# 2. 입력 변수 구성
# Segment와 ID, 혹시 제거 안 된 텍스트형 제외
X = df10.drop(columns=[target])
y = df10[target]

# 3. 필요 시 범주형을 숫자로 변환 (예: Label Encoding)
# object 타입이 있다면 처리 필요
for col in X.select_dtypes(include='object').columns:
    X[col] = LabelEncoder().fit_transform(X[col])

# 4. 목표변수 라벨 인코딩
y_encoded = LabelEncoder().fit_transform(y)

# 5. 학습/검증 데이터 분리
X_train, X_valid, y_train, y_valid = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# 6. 랜덤포레스트 학습
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 7. 변수 중요도 시각화
importances = pd.Series(model.feature_importances_, index=X.columns)
top_importances = importances.sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_importances.values, y=top_importances.index, palette='viridis')
plt.title("🎯 Top 20 Feature Importances (Random Forest)")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

```python
# 중요도 기준 상위 N개 변수 리스트 추출
top_features = importances.sort_values(ascending=False).head(20).index.tolist()

# 수치형 변수만 필터링
numeric_features = df10[top_features].select_dtypes(include=['int64', 'float64']).columns.tolist()

print("🔍 상위 중요도 수치형 변수 목록:")
print(numeric_features)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Segment 더미 인코딩
df10_dummied = pd.get_dummies(df10, columns=['Segment'], prefix='Segment')

# 2. 중요도 상위 변수 20개 추출
top_features = importances.sort_values(ascending=False).head(20).index.tolist()

# 3. 수치형 변수만 필터링 + Segment 더미 컬럼 추가
numeric_features = df10_dummied[top_features].select_dtypes(include=['int64', 'float64']).columns.tolist()

# Segment 더미 컬럼 추가 (A ~ E 기준 가정)
segment_dummies = [col for col in df10_dummied.columns if col.startswith('Segment_')]
numeric_features += segment_dummies

# 4. 상관계수 히트맵
corr_matrix = df10_dummied[numeric_features].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, vmax=1, vmin=-1)
plt.title('📌 중요 수치형 변수 + Segment 더미 변수 상관관계 Heatmap')
plt.tight_layout()
plt.show()
```
