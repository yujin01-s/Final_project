```python
import pandas as pd
import numpy as np

# â¬…ï¸ ì‚¬ìš©í•  ì „ì²´ ì»¬ëŸ¼ ëª©ë¡ì— ë¹„ìœ¨ ê³„ì‚°ìš© ì»¬ëŸ¼ë„ í¬í•¨ì‹œì¼œ ë¨¼ì € ë¶ˆëŸ¬ì™€ì•¼ í•¨
file_path = 'data/í†µí•©_train_ë³‘í•©ì™„ë£Œ.parquet'

# âœ… train ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
train_df = pd.read_parquet(file_path)
```

```python
def map_categorical_columns(df, verbose=True):
    """
    ë¯¸ë¦¬ ì •ì˜ëœ ë§¤í•‘ ê¸°ì¤€ì— ë”°ë¼ ë²”ì£¼í˜• ì»¬ëŸ¼ë“¤ì„ ìˆ˜ì¹˜í˜•ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
    ì²˜ë¦¬ ì»¬ëŸ¼: ê±°ì£¼ì‹œë„ëª…, ì—°íšŒë¹„ë°œìƒì¹´ë“œìˆ˜_B0M, í•œë„ì¦ì•¡íšŸìˆ˜_R12M, ì´ìš©ê¸ˆì•¡ëŒ€,
              í• ì¸ê±´ìˆ˜_R3M, í• ì¸ê±´ìˆ˜_B0M, ë°©ë¬¸íšŸìˆ˜_PC_R6M, ë°©ë¬¸íšŸìˆ˜_ì•±_R6M, ë°©ë¬¸ì¼ìˆ˜_PC_R6M
    """

    # 1. ê±°ì£¼ì‹œë„ëª… â†’ ìˆ˜ë„ê¶Œ ì—¬ë¶€
    capital_area = ['ì„œìš¸íŠ¹ë³„ì‹œ', 'ê²½ê¸°ë„', 'ì¸ì²œê´‘ì—­ì‹œ']
    if 'ê±°ì£¼ì‹œë„ëª…' in df.columns:
        df['ê±°ì£¼ì‹œë„ëª…'] = df['ê±°ì£¼ì‹œë„ëª…'].apply(lambda x: 1 if x in capital_area else 0)
        if verbose: print("[ê±°ì£¼ì‹œë„ëª…] â†’ ìˆ˜ë„ê¶Œ ì—¬ë¶€ ì¸ì½”ë”© ì™„ë£Œ")

    # 2. ì—°íšŒë¹„ë°œìƒì¹´ë“œìˆ˜_B0M
    mapping = {"0ê°œ": 0, "1ê°œì´ìƒ": 1}
    if 'ì—°íšŒë¹„ë°œìƒì¹´ë“œìˆ˜_B0M' in df.columns:
        df['ì—°íšŒë¹„ë°œìƒì¹´ë“œìˆ˜_B0M'] = df['ì—°íšŒë¹„ë°œìƒì¹´ë“œìˆ˜_B0M'].map(mapping).astype(int)
        if verbose: print("[ì—°íšŒë¹„ë°œìƒì¹´ë“œìˆ˜_B0M] ì¸ì½”ë”© ì™„ë£Œ")

    # 3. í•œë„ì¦ì•¡íšŸìˆ˜_R12M
    mapping = {"0íšŒ": 0, "1íšŒì´ìƒ": 1}
    if 'í•œë„ì¦ì•¡íšŸìˆ˜_R12M' in df.columns:
        df['í•œë„ì¦ì•¡íšŸìˆ˜_R12M'] = df['í•œë„ì¦ì•¡íšŸìˆ˜_R12M'].map(mapping).astype(int)
        if verbose: print("[í•œë„ì¦ì•¡íšŸìˆ˜_R12M] ì¸ì½”ë”© ì™„ë£Œ")

    # 4. ì´ìš©ê¸ˆì•¡ëŒ€ (ì¤‘ê°„ê°’ ê¸°ì¤€: ë§Œì› ë‹¨ìœ„)
    mapping = {
        "09.ë¯¸ì‚¬ìš©": 0,
        "05.10ë§Œì›-": 5,
        "04.10ë§Œì›+": 20,
        "03.30ë§Œì›+": 40,
        "02.50ë§Œì›+": 75,
        "01.100ë§Œì›+": 150
    }
    if 'ì´ìš©ê¸ˆì•¡ëŒ€' in df.columns:
        df['ì´ìš©ê¸ˆì•¡ëŒ€'] = df['ì´ìš©ê¸ˆì•¡ëŒ€'].map(mapping)
        if verbose: print("[ì´ìš©ê¸ˆì•¡ëŒ€] ì¤‘ê°„ê°’ ì¸ì½”ë”© ì™„ë£Œ")

    # 5. í• ì¸ê±´ìˆ˜ ì¸ì½”ë”©
    discount_map = {
        "1íšŒ ì´ìƒ": 1,
        "10íšŒ ì´ìƒ": 10,
        "20íšŒ ì´ìƒ": 20,
        "30íšŒ ì´ìƒ": 30,
        "40íšŒ ì´ìƒ": 40
    }
    for col in ['í• ì¸ê±´ìˆ˜_R3M', 'í• ì¸ê±´ìˆ˜_B0M']:
        if col in df.columns:
            df[col] = df[col].map(discount_map).astype(int)
            if verbose: print(f"[{col}] ì¸ì½”ë”© ì™„ë£Œ")

    # 6. ë°©ë¬¸íšŸìˆ˜ ë° ë°©ë¬¸ì¼ìˆ˜ ì¸ì½”ë”©
    visit_map = {
        "1íšŒ ì´ìƒ": 1,
        "10íšŒ ì´ìƒ": 10,
        "20íšŒ ì´ìƒ": 20,
        "30íšŒ ì´ìƒ": 30,
        "40íšŒ ì´ìƒ": 40,
        "50íšŒ ì´ìƒ": 50,
        "60íšŒ ì´ìƒ": 60,
        "70íšŒ ì´ìƒ": 70,
        "80íšŒ ì´ìƒ": 80
    }

    visit_cols = ['ë°©ë¬¸íšŸìˆ˜_PC_R6M', 'ë°©ë¬¸íšŸìˆ˜_ì•±_R6M', 'ë°©ë¬¸ì¼ìˆ˜_PC_R6M']
    for col in visit_cols:
        if col in df.columns:
            df[col] = df[col].map(visit_map).astype(int)
            if verbose: print(f"[{col}] ì¸ì½”ë”© ì™„ë£Œ")

    return df
```

```python
# ğŸ“Œ í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd
import numpy as np
import optuna
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from imblearn.over_sampling import RandomOverSampler

# ğŸ“Œ 1. ë°ì´í„° ì „ì²˜ë¦¬
X = train_df.drop(columns=["ID", "Segment"]).copy()
y = train_df["Segment"]

# ë²”ì£¼í˜• ì¸ì½”ë”©
cat_cols = X.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))

# ìŠ¤ì¼€ì¼ë§
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# íƒ€ê²Ÿ ì¸ì½”ë”©
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)

# ğŸ“Œ 2. Optunaë¡œ í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹ (3-Fold êµì°¨ê²€ì¦ ê¸°ì¤€)
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 180, 350),
        'max_depth': trial.suggest_int('max_depth', 7, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.05, 0.1),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'tree_method': 'gpu_hist',
        'predictor': 'gpu_predictor',
        'use_label_encoder': False,
        'eval_metric': 'mlogloss',
        'random_state': 42
    }

    model = XGBClassifier(**params)

    # ê°„ë‹¨í•œ í‰ê°€ (K-Fold ì—†ì´ ì „ì²´ì—ì„œ cross_val)
    score = cross_val_score(model, X_scaled, y_encoded, cv=3, scoring='f1_weighted')
    return score.mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)  # âœ… ì»´í“¨í„° ë¦¬ì†ŒìŠ¤ ê³ ë ¤í•´ì„œ 20íšŒë§Œ

# ìµœì  íŒŒë¼ë¯¸í„° í™•ì¸
best_params = study.best_params
print("\nâœ… ìµœì  í•˜ì´í¼íŒŒë¼ë¯¸í„°:")
for k, v in best_params.items():
    print(f"{k}: {v}")
```

```python
# ğŸ“Œ 3. K-Fold í•™ìŠµ (íŠœë‹ëœ íŒŒë¼ë¯¸í„°ë¡œ)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
ros = RandomOverSampler(sampling_strategy={0: 2000, 1: 2000}, random_state=42)

fold = 1
all_reports = []

for train_idx, val_idx in skf.split(X_scaled, y_encoded):
    X_train, X_val = X_scaled.iloc[train_idx], X_scaled.iloc[val_idx]
    y_train, y_val = y_encoded[train_idx], y_encoded[val_idx]

    X_train_os, y_train_os = ros.fit_resample(X_train, y_train)
    print(f"\nğŸ” Fold {fold} ì˜¤ë²„ìƒ˜í”Œë§ ì™„ë£Œ:", pd.Series(y_train_os).value_counts().to_dict())

    # íŠœë‹ëœ íŒŒë¼ë¯¸í„° + ê³ ì • íŒŒë¼ë¯¸í„° êµ¬ì„±
    xgb_model = XGBClassifier(
        **best_params,
        tree_method='gpu_hist',
        predictor='gpu_predictor',
        use_label_encoder=False,
        eval_metric='mlogloss',
        random_state=42
    )

    xgb_model.fit(X_train_os, y_train_os)
    y_pred = xgb_model.predict(X_val)

    print(f"\nğŸ“˜ Fold {fold} ê²°ê³¼")
    print(classification_report(y_val, y_pred, target_names=le_y.classes_))

    all_reports.append(classification_report(y_val, y_pred, output_dict=True))
    fold += 1
```

```python
import optuna
from sklearn.metrics import f1_score
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# âœ… í™•ë¥  ì˜ˆì¸¡ (ì—†ì„ ê²½ìš° ì¶”ê°€)
# X_valê³¼ xgb_model, y_valì´ ì´ë¯¸ ìˆì–´ì•¼ í•¨
y_val_proba = xgb_model.predict_proba(X_val)

# âœ… ê²°ê³¼ ì €ì¥ ë¦¬ìŠ¤íŠ¸
best_thresholds = []
best_f1s = []

# âœ… í´ë˜ìŠ¤ ìˆ˜ ìë™ ê°ì§€
n_classes = y_val_proba.shape[1]

# âœ… í´ë˜ìŠ¤ë³„ë¡œ ë°˜ë³µ ì‹¤í–‰
for class_idx in range(n_classes):
    print(f"ğŸ” í´ë˜ìŠ¤ {class_idx} ìµœì  threshold íƒìƒ‰ ì¤‘...")

    target_class = class_idx  # ì™¸ë¶€ì—ì„œ objective í•¨ìˆ˜ê°€ ì°¸ì¡°í•¨

    def objective(trial):
        threshold = trial.suggest_float("threshold", 0.1, 0.9)
        y_pred = []
        for row in y_val_proba:
            if row[target_class] >= threshold:
                y_pred.append(target_class)
            else:
                y_pred.append(np.argmax(row))
        return f1_score(y_val, y_pred, average='weighted')

    study = optuna.create_study(direction="maximize", sampler=optuna.samplers.TPESampler(seed=42))
    study.optimize(objective, n_trials=50)

    best_thresh = study.best_params["threshold"]
    best_f1 = study.best_value

    best_thresholds.append(best_thresh)
    best_f1s.append(best_f1)

    print(f"âœ… Class {class_idx}: Threshold = {best_thresh:.2f}, F1 = {best_f1:.4f}")

# âœ… ìµœì¢… í™•ì¸
print("\nğŸ¯ í´ë˜ìŠ¤ë³„ Best Threshold ë¦¬ìŠ¤íŠ¸:")
for idx, t in enumerate(best_thresholds):
    print(f"  Class {idx}: {t:.2f}")
```

```python
# âœ… ê²°ê³¼ ì‹œê°í™”
df_thresh = pd.DataFrame({
    "Segment": le_y.inverse_transform(list(range(n_classes))),
    "Best_Threshold": best_thresholds,
    "Best_F1": best_f1s
})

fig, ax1 = plt.subplots(figsize=(10, 6))

# Threshold ë§‰ëŒ€
sns.barplot(data=df_thresh, x="Segment", y="Best_Threshold", color="skyblue", ax=ax1)
ax1.set_ylabel("Best Threshold", color="skyblue")
ax1.tick_params(axis='y', labelcolor="skyblue")

# F1 ì„  ê·¸ë˜í”„ ì¶”ê°€
ax2 = ax1.twinx()
sns.lineplot(data=df_thresh, x="Segment", y="Best_F1", color="orange", marker="o", ax=ax2)
ax2.set_ylabel("Best F1 Score", color="orange")
ax2.tick_params(axis='y', labelcolor="orange")

plt.title("Segmentë³„ Best Threshold + F1 Score (Optuna)")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report

# âœ… ë¨¼ì € threshold ê¸°ë°˜ ì˜ˆì¸¡ í•¨ìˆ˜ ì •ì˜
def predict_with_thresholds(y_proba, thresholds, exclude_from_fallback=None):
    preds = []
    for row in y_proba:
        candidates = [i for i, p in enumerate(row) if p >= thresholds[i]]
        if candidates:
            preds.append(candidates[np.argmax([row[i] for i in candidates])])
        else:
            row_copy = row.copy()
            if exclude_from_fallback is not None:
                for idx in exclude_from_fallback:
                    row_copy[idx] = -1
            preds.append(np.argmax(row_copy))
    return np.array(preds)

# âœ… threshold ì ìš© ì˜ˆì¸¡
y_val_pred_encoded = predict_with_thresholds(y_val_proba, best_thresholds)
y_val_pred = le_y.inverse_transform(y_val_pred_encoded)
y_val_true = le_y.inverse_transform(y_val)

# âœ… classification report ì¶œë ¥
print("\nğŸ“Š Classification Report (Best Threshold ì ìš©):")
print(classification_report(y_val_true, y_val_pred, target_names=le_y.classes_))
```

```python
from datetime import datetime


# âœ… 2. í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° (ë³‘í•© ë° ì „ì²˜ë¦¬ëœ ìƒíƒœë¼ê³  ê°€ì •)
test_df = pd.read_parquet("data/í†µí•©_test_ë³‘í•©ì™„ë£Œ.parquet")

# âœ… 3. í”¼ì²˜ ì„ íƒ
all_features = list(imputer.feature_names_in_)  # í•™ìŠµ ì‹œ ì‚¬ìš©í•œ í”¼ì²˜ ìˆœì„œ ê·¸ëŒ€ë¡œ
# ì…ë ¥ ë°ì´í„° êµ¬ì„±
X_test = test_df[all_features].copy()

# âœ… 4. ë²”ì£¼í˜• ì¸ì½”ë”© (1ì°¨: ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜)
X_test = map_categorical_columns(X_test, verbose=False)

# âœ… 4-1. ë‚¨ì€ object ì»¬ëŸ¼ì´ ìˆë‹¤ë©´ ì¶”ê°€ ì¸ì½”ë”©
remaining_obj_cols = X_test.select_dtypes(include='object').columns.tolist()
if remaining_obj_cols:
    print(f"ğŸš¨ ì¶”ê°€ ì¸ì½”ë”© ëŒ€ìƒ ì»¬ëŸ¼: {remaining_obj_cols}")
    for col in remaining_obj_cols:
        le = LabelEncoder()
        X_test[col] = le.fit_transform(X_test[col].astype(str))


# âœ… 5. ê²°ì¸¡ì¹˜ ì²˜ë¦¬ ë° ìŠ¤ì¼€ì¼ë§
X_test = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)
X_test_scaled = scaler.transform(X_test)

# âœ… 6. ì˜ˆì¸¡ í™•ë¥  ê³„ì‚°
y_test_proba = xgb_model.predict_proba(X_test_scaled)

# âœ… 7. Threshold ê¸°ë°˜ ì˜ˆì¸¡ í•¨ìˆ˜
def predict_with_thresholds(y_proba, thresholds, exclude_from_fallback=None):
    preds = []
    for row in y_proba:
        candidates = [i for i, p in enumerate(row) if p >= thresholds[i]]
        if candidates:
            preds.append(candidates[np.argmax([row[i] for i in candidates])])
        else:
            row_copy = row.copy()
            if exclude_from_fallback is not None:
                for idx in exclude_from_fallback:
                    row_copy[idx] = -1
            preds.append(np.argmax(row_copy))
    return np.array(preds)

# âœ… 8. Threshold ì ìš© ì˜ˆì¸¡
a_index = le_y.transform(['A'])[0]

y_test_pred_encoded = predict_with_thresholds(
    y_test_proba, best_thresholds, exclude_from_fallback=[a_index]
)
y_test_pred = le_y.inverse_transform(y_test_pred_encoded)

# âœ… 9. ê²°ê³¼ ì €ì¥
submission = pd.DataFrame({
    'ID': test_df['ID'],
    'Segment': y_test_pred
})
submission = submission.drop_duplicates(subset='ID', keep='first')

today = datetime.now().strftime('%Y%m%d')
submission.to_csv(f"submission_{today}.csv", index=False)
print(f"âœ… ì˜ˆì¸¡ ì™„ë£Œ ë° ì €ì¥: submission_{today}.csv")
```

```python
import numpy as np
import pandas as pd

# SHAP í¸ì°¨ ê³„ì‚° (axis=2 â†’ ì„¸ê·¸ë¨¼íŠ¸ë³„ SHAP í¸ì°¨, mean(axis=1) â†’ ì „ì²´ í”¼ì²˜ í‰ê· )
shap_std_matrix = np.std(shap_values.values, axis=2)        # shape: (n_samples, n_features)
shap_std_mean = np.mean(shap_std_matrix, axis=1)             # shape: (n_samples, )

# ì˜ˆì¸¡ í™•ë¥  (ê°€ì¥ ë†’ì€ í™•ë¥ )
proba = xgb_model.predict_proba(X_shap_sample)
max_proba = np.max(proba, axis=1)
prob_uncertainty = 1 - np.abs(0.5 - max_proba)  # í™•ë¥ ì´ 0.5ì¼ìˆ˜ë¡ ë¶ˆí™•ì‹¤ì„±â†‘

# ìŠ¤ì½”ì–´ë§ (ê°€ì¤‘ì¹˜ëŠ” ìƒí™©ì— ë”°ë¼ ì¡°ì • ê°€ëŠ¥)
score = shap_std_mean + 0.5 * prob_uncertainty

# ê²°ê³¼ ì •ë¦¬
score_df = pd.DataFrame({
    'Score': score,
    'Max Proba': max_proba,
    'Prob Uncertainty': prob_uncertainty,
    'SHAP Std Mean': shap_std_mean
}, index=X_shap_sample.index)

# ìƒìœ„ ì „í™˜ ë¯¼ê° ê³ ê° ì¶”ì¶œ
top_candidates = score_df.sort_values(by='Score', ascending=False).head(10)
display(top_candidates)
```
