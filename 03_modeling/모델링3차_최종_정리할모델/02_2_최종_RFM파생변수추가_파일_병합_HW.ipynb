```python
# 테스트 데이터 컬럼
base_test_cols = ['ID']
all_selected_test_cols= list(set(base_test_cols + all_features))
len(all_selected_test_cols)
```

```python
import pandas as pd

# ✅ 1. 원본 train 데이터 선택 컬럼만 불러오기
train_df = pd.read_parquet("data/통합_train_데이터.parquet", columns=all_selected_train_cols)

# ✅ 2. 추가 컬럼 불러오기 (R_score, F_score, M_score)
extra_df = pd.read_csv("data/rfm_only.csv")

# ✅ 3. (ID, Segment) 기준 중복 제거
extra_df = extra_df.drop_duplicates(subset=["ID", "Segment"])

# ✅ 4. 병합 (ID + Segment 기준)
train_df = train_df.merge(extra_df, on=["ID", "Segment"], how="left")

# ✅ 5. 파생 변수 생성: 이용카드비율_신용
if '이용카드수_신용' in train_df.columns and '유효카드수_신용' in train_df.columns:
    train_df['이용카드비율_신용'] = train_df.apply(
        lambda row: row['이용카드수_신용'] / row['유효카드수_신용']
        if row['유효카드수_신용'] != 0 else 0,
        axis=1
    )

# ✅ 6. 저장
train_df.to_parquet("data/통합_train_병합완료.parquet", index=False)
print("✅ 병합 및 저장 완료: 통합_train_병합완료.parquet (중복 제거 포함)")
```

```python
import pandas as pd

# ✅ 1. 원본 테스트 데이터 불러오기
test_df = pd.read_parquet("data/통합_test_데이터.parquet", columns=all_selected_test_cols)
test_df['ID']
```

```python

# ✅ 2. 추가 컬럼 불러오기 (R_score, F_score, M_score 포함)
extra_df = pd.read_csv("data/rfm_test_only.csv")

# ✅ 3. (ID, Segment) 기준 중복 제거 → 이후 ID 기준 병합 가능하게 만듦
extra_df = extra_df.drop_duplicates(subset=["ID"])

extra_df
```

```python
# ✅ 4. Segment 기준 제거하고 ID 기준 병합 준비
test_df = test_df.merge(extra_df, on="ID", how="left")

# ✅ 5. 파생 변수 생성: 이용카드비율_신용
if '이용카드수_신용' in test_df.columns and '유효카드수_신용' in test_df.columns:
    test_df['이용카드비율_신용'] = test_df.apply(
        lambda row: row['이용카드수_신용'] / row['유효카드수_신용']
        if row['유효카드수_신용'] != 0 else 0,
        axis=1
    )

test_df
```

```python
# ✅ 6. 저장
test_df.to_parquet("data/통합_test_병합완료.parquet", index=False)
print("✅ 병합 및 저장 완료: 통합_test_병합완료.parquet (ID 기준, Segment 제거됨)")
```
