```python
import pandas as pd

# 1. Parquet 파일 불러오기
file_path = "../../data/merged/성과정보_train.parquet"
df = pd.read_parquet(file_path)

# 2. 키워드가 포함된 컬럼만 필터링
keywords = ["잔액_신판"]
bal_add = [col for col in df.columns if any(k in col for k in keywords)]

# 3. 선택된 컬럼만 추출한 데이터프레임 생성
df_selected = df[bal_add].copy()

# 4. 결과 출력
print(f"선택된 컬럼 수: {len(bal_add)}")
print("컬럼 목록:", bal_add)
print(df_selected.head())
```
