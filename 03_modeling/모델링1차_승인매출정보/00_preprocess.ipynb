```python
import pandas as pd
import os

# 파일 경로
base_path = '../../data/train/3.승인매출정보'

# 201807~201812까지 불러오기
months = ['201807', '201808', '201809', '201810', '201811', '201812']
승인매출정보_list = []

for month in months:
    file_path = os.path.join(base_path, f'{month}_train_승인매출정보.parquet')
    df = pd.read_parquet(file_path)
    df['기준년월'] = month  # 기준년월 컬럼 추가
    승인매출정보_list.append(df)

# 병합
승인매출정보 = pd.concat(승인매출정보_list, ignore_index=True)

# 확인
print("승인매출정보 shape:", 승인매출정보.shape)
print("컬럼 확인:", 승인매출정보.columns.tolist())
```

```python
승인매출정보.to_csv('merged_train_data.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import os

# 디렉토리 경로 설정
base_path = '../../data/train/1.회원정보'
months = ['201807', '201808', '201809', '201810', '201811', '201812']

회원정보_list = []

for month in months:
    file_path = os.path.join(base_path, f'{month}_train_회원정보.parquet')
    df = pd.read_parquet(file_path)
    print(f"{month} 불러오기 완료, shape: {df.shape}, 컬럼: {df.columns.tolist()}")
    회원정보_list.append(df)

# 모든 월 데이터 병합
회원정보_all = pd.concat(회원정보_list, ignore_index=True)

# 중복 ID 제거 및 확인
print("전체 회원정보 shape:", 회원정보_all.shape)
print("고유 ID 수:", 회원정보_all['ID'].nunique())
```

```python
# ID와 Segment만 추출하고 중복 제거
segment_info = 회원정보_all[['ID', 'Segment']].drop_duplicates(subset='ID')
print("Segment 정보 shape:", segment_info.shape)
```

```python
승인매출정보_with_segment = 승인매출정보.merge(segment_info, on='ID', how='left')
```

```python
# 병합 후 데이터 개수 및 Segment 분포 확인
print("병합 후 shape:", 승인매출정보_with_segment.shape)
print("Segment 값 분포:\n", 승인매출정보_with_segment['Segment'].value_counts(dropna=False))
```

```python
승인매출정보_with_segment['Segment'].isna().sum()
```

```python
# CSV로 저장 (엑셀에서도 열 수 있도록 utf-8-sig 인코딩)
승인매출정보_with_segment.to_csv('test_승인매출정보_segment.csv', index=False, encoding='utf-8-sig')
```
