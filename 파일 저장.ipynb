```python
import pandas as pd
import glob
import os

# ì‹œíŠ¸ ë²ˆí˜¸ ë° ì •í™•í•œ í´ë”ëª… ë§¤í•‘ (ë„ì–´ì“°ê¸° ì œê±° + ê²½ë¡œ ë°˜ì˜)
table_paths = {
    "1.íšŒì›ì •ë³´": "data/test/1.íšŒì›ì •ë³´",
    "2.ì‹ ìš©ì •ë³´": "data/test/2.ì‹ ìš©ì •ë³´",
    "3.ìŠ¹ì¸ë§¤ì¶œì •ë³´": "data/test/3.ìŠ¹ì¸ë§¤ì¶œì •ë³´",
    "4.ì²­êµ¬ì…ê¸ˆì •ë³´": "data/test/4.ì²­êµ¬ì…ê¸ˆì •ë³´",
    "5.ì”ì•¡ì •ë³´": "data/test/5.ì”ì•¡ì •ë³´",
    "6.ì±„ë„ì •ë³´": "data/test/6.ì±„ë„ì •ë³´",
    "7.ë§ˆì¼€íŒ…ì •ë³´": "data/test/7.ë§ˆì¼€íŒ…ì •ë³´",
    "8.ì„±ê³¼ì •ë³´": "data/test/8.ì„±ê³¼ì •ë³´"
}

# ê²°ê³¼ ì €ì¥ìš© ë”•ì…”ë„ˆë¦¬
merged_data = {}

# ê° í…Œì´ë¸” í†µí•©
for name, path in table_paths.items():
    files = glob.glob(os.path.join(path, "*.parquet"))
    if len(files) > 0:
        df = pd.concat([pd.read_parquet(f) for f in sorted(files)], ignore_index=True)
        merged_data[name] = df
        print(f"{name} âœ… ë¶ˆëŸ¬ì˜¤ê¸° ì™„ë£Œ: {df.shape}")
    else:
        print(f"{name} âš ï¸ íŒŒì¼ ì—†ìŒ")
```
